```python
# Tema pe acasă 1.1
# Ecuația de gradul 2
```

```python
import math
a = int(input("a = "))
b = int(input("b = "))
c = int(input("c = "))
delta = b ** 2 - (4 * a * c)
if delta < 0:
    print("Ecuația nu are soluții")
if delta == 0:
    x = -b / (2 * a)
    print(f"Soluția este: x1 = x2 = {x}")
elif delta > 0:
    x1 = (-b + math.sqrt(delta)) / (2 * a)
    x2 = (-b - math.sqrt(delta)) / (2 * a)
    print(f"Soluția este: x1 = {x1}, x2 = {x2}")
```

```python
# tema pentru acasă 1.2
# calculează distanța euclidiană: d(x, y)
```

```python
import math
x1 = int(input("x1 = "))
x2 = int(input("x2 = "))
y1 = int(input("y1 = "))
y2 = int(input("y2 = "))
res = math.sqrt((x1-x2)**2 +(y1-y2)**2)
print(f"d(x, y) = {res}")
```
